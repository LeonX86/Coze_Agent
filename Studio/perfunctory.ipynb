```python
with open('./fy_dict.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

fy_dict = []
for line in lines:
    # 分割行号和内容，取|后面的部分
    content = line.strip().split('|', 1)[1] if '|' in line else line.strip()
    fy_dict.append(content)

# 打印成可复制的列表格式
print("fy_dict = [")
for i, item in enumerate(fy_dict):
    if i < len(fy_dict) - 1:
        print(f"    '{item}',")
    else:
        print(f"    '{item}'")
print("]")
```
